# Changing pandas Options with Attributes and Dot Syntax (section 14.173):

```python
# fake dataset:
data = np.random.randint(0, 100, [1000, 50])
# this will generate a dataset of random numbers from 0-100 with  999 rows and 49 columns
df = pd.DataFrame(data)
df.tail(3)
```

```python
pd.options.display.max_rows
# this is the max number of rows pandas will output visually
pd.options.display.max_rows = 6
# can set how many are outputted
df
```

```python
pd.options.display.max_columns
```

```python
pd.options.display.max_columns = 6
df
```

# Changing Pandas Options with Methods (section 14.174):

- alternate syntax for the previous lesson
- called directly on pandas

```python
# fake dataset:
data = np.random.randint(0, 100, [1000, 50])
# this will generate a dataset of random numbers from 0-100 with  999 rows and 49 columns
df = pd.DataFrame(data)
df.tail(3)
```

```python
pd.get_option("max_rows")
```

```python
pd.get_option("max_columns")
```

```python
pd.set_option("max_rows", 4)
df
```

```python
pd.set_option("max_columns", 10)
df
```

```python
pd.reset_option("max_rows")
pd.get_option("max_rows")
```

```python
pd.reset_option("max_columns")
pd.get_option("max_columns")
```

```python
pd.describe_option("max_columns")
```

```python
pd.describe_option("max_rows")
```

# The Precision Option (section 14.175):

- alters the number of digits that appear after the decimal point in a float
- doesn't actually modify the number, just how it is displayed
- pandas defaults to 6 decimal places in floats

```python
# fake random df with floats:
df = pd.DataFrame(np.random.randn(5, 5))
df
```

```python
pd.get_option("precision")
```

```python
pd.set_option("precision", 2)
df
```

```python
pd.reset_option("precision")
df
# the original numbers have been preserved!
```
