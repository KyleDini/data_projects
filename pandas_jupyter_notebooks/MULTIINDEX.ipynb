# Exctract Index Level Values with .get_level_values() method (section 8.112):

- used on an index of a multiindex df
- extracts all of the values from a level in the multiindex
- good to check for inclusion in a multiindex and want to specify the level you want to check in

```python
bigmac = pd.read_csv("bigmac.csv", parse_dates = ["Date"], index_col = ["Date", "Country"])
bigmac.sort_index(inplace = True)
bigmac.head(3)
# provide a list to create a multiindex when importing
```

```python
bigmac.index.get_level_values("Date")
bigmac.index.get_level_values(0)
```

```python
bigmac.index.get_level_values("Country")
bigmac.index.get_level_values(1)
```

# Change Index Level Name with .set_names() method (section 8.113):

- invoked on the index object, not the whole df

```python
bigmac = pd.read_csv("bigmac.csv", parse_dates = ["Date"], index_col = ["Date", "Country"])
bigmac.sort_index(inplace = True)
bigmac.head(3)
```

# The .sort_index() method on a MultiIndex DataFrame (section 8.114):

```python
bigmac = pd.read_csv("bigmac.csv", parse_dates = ["Date"], index_col = ["Date", "Country"])
bigmac.head(3)
```

```python
bigmac.sort_index()
bigmac.sort_index(ascending = False)
# can pass ascending a *list* of booleans to customize the ascending for each level, goes in order of index
bigmac.sort_index(ascending = [True, False])
bigmac.sort_index(ascending = [False, True], inplace = True)
bigmac.head()
```

```python
bigmac.sort_index(level = 0)
# tells pandas that we only want to sort by the specified level of the index, ignore the sorting of the other ones
# keeps the second level in original order (the countries were already alphabetized in the og csv)
bigmac.sort_index(level = 1)
bigmac.sort_index(level = "Country", ascending = False)
# generally better to sort all the levels , but depends on each case
```

# Extract Rows from a MultiIndex DataFrame (section 8.115):

- use .loc[] for index labels
- use .iloc[] for index positions

```python
bigmac = pd.read_csv("bigmac.csv", parse_dates = ["Date"], index_col = ["Date", "Country"])
bigmac.sort_index(inplace = True)
bigmac.head(3)
```

```python
# the second arguement for .loc[] can be either a value from the second level of the index, or a column name
# ^ this is ambiguous, when calling one other the other you dont know which type is being returned
# pandas documentation recommends having the first argument have all row info, second has all column info
# recommended syntax shown below (feed info in as a tuple, not a list):

bigmac.loc[("2010-01-01", "Argentina")]
bigmac.loc[("2010-01-01", "Argentina"), "Price in US Dollars"]
# if we wanted to access more columns, we would put them in a list**
bigmac.loc[("2010-01-01", "Argentina"), ["Price in US Dollars", "Price in US Dollars"]]
```

```python
# need to place a comma inside of a tuple with just one element for python to recognize it as a tuple
bigmac.loc[("2010-01-01",)].head()
```

```python
# multiindexes don't affect iloc[] at all because each row is still given an index position in line
bigmac.iloc[13]
# use a list to extract multiple rows with index position
bigmac.iloc[[2, 5, 44]]
```

# The .transpose() Method (section 8.116):

- swaps the indexes of the df
- takes the row labels and moves them to be the column headers, takes column headers and moves them to be the row labels
- .transpose() doesn't mutate a df, doesn't include an inplace parameter
- need to assign it to a variable to preserve it

```python
bigmac = pd.read_csv("bigmac.csv", parse_dates = ["Date"], index_col = ["Date", "Country"])
bigmac.sort_index(inplace = True)
bigmac.head(3)
```

```python
bigmac = bigmac.transpose()
# created a column multiindex in this example
# this multiindex has two column levels: date and country
bigmac.head()
```

```python
# good practice to wrap first arguement of .loc[] in a tuple when working with a multiindex df, even if only one value
bigmac.loc[("Price in US Dollars",)]
# can give .loc[] a tuple that specifies the values from each respective level in the column mutliindex
bigmac.loc[("Price in US Dollars",), ("2010-01-01",)]
bigmac.loc[("Price in US Dollars",), ("2010-01-01", "China")]
```

```python
bigmac.loc[("Price in US Dollars",), ("2010-01-01", "China"):("2010-01-01", "Egypt")]
# pulls china through egypt price in usd
# multiindex rows or columns need multiple reference points
```

# The .swaplevel() method (section 8.117): 

- swaps to index levels in a multiindex
- .swaplevel() is given arguements that represent which levels you want to swap
- very helpful for multiindexes with 3 or more levels
- doesn't include an inplace parameter

```python
bigmac = pd.read_csv("bigmac.csv", parse_dates = ["Date"], index_col = ["Date", "Country"])
bigmac.sort_index(inplace = True)
bigmac.head(3)
```

```python
bigmac.swaplevel("Date", "Country")
bigmac.swaplevel("Country", "Date")
# no need to provide additional arguements if you only have two levels
# argument order doesn't matter
bigmac.swaplevel(0, 1)
bigmac.swaplevel(1, 0)
bigmac.swaplevel()
# all of these lines of code do the same thing in this example
```

```python
bigmac = bigmac.swaplevel()
bigmac.head(3)
```

# The .stack() method (section 8.118):

- takes the column index and moves it to the horizontal index (makes more sense when you see it)
- columns are techincally also an index, but are not what someone refers to when they say index
- doesn't change the contents of the df, just shifts how we see it

```python
world = pd.read_csv("worldstats.csv", index_col = ["country", "year"])
world.head(3)
```

```python
world.stack()
# this example effectively doubled the number of rows, and halved the number of columns
# since this example leaves pandas with one column of data, it has converted our df into a series
world.stack().to_frame()
# to_frame() keeps the series as a df
```

# The .unstack() method, part 1 (section 8.119):

- does the reverse of the .stack() method
- goes to the deepest index within the multiindex and moves it to the columns axis
- can combine stack and unstack until you get it to the shape that you want it

```python
world = pd.read_csv("worldstats.csv", index_col = ["country", "year"])
world.head(3)
```

```python
s = world.stack()

s.unstack()

s.unstack().unstack().unstack()
# this example created a three layer series, but in reverse of what .stack() created
```

# The .unstack() method, part 2 (section 8.120):

- some arguements you can provide to the .unstack() method shown below:

```python
world = pd.read_csv("worldstats.csv", index_col = ["country", "year"])
world.head(3)
```

```python
s = world.stack()
s.head(3)
```

```python
# can provide the index that we want to move from our mutliindex, each mi layer has its own index position
s.unstack(1)
```

```python
# can also provide numeric arguements
s.unstack(-2)
# is the same as:
s.unstack(1)
```

```python
# can also write the string name of the index if there is one
s.unstack("year")
# or
s.unstack("country")
```

# The .unstack() method, part 3 (section 8.121):

- how to unstack multiple layers of a multiindex to the columns axis at once:

```python
world = pd.read_csv("worldstats.csv", index_col = ["country", "year"])
s = world.stack()
s.head(3)
```

```python
s.unstack(level = [0, 1])
s.unstack(["country", "year"])
```

```python
s = s.unstack("year", fill_value = "Null Value")
s.head(3)
```

# The pivot method (section 8.122):

- Used to reorient a dataset, takes the values currently in a column and converts them to column headers

```python
sales = pd.read_csv("salesmen.csv", parse_dates = ["Date"])
sales["Salesman"] = sales["Salesman"].astype("category")
sales.head(3)
```

```python
sales["Salesman"].value_counts()
# this data would be more readable if the salesman names were the column headers, and put the revenue as the values
```

```python
sales.pivot(index = "Date", columns = "Salesman", values = "Revenue")
```

# The pivot_table() method (section 8.123):

- ideal for aggregating a dataset to look at it as a whole

```python
foods = pd.read_csv("foods.csv")
foods.head(3)
```

```python
foods.pivot_table(values = "Spend", index = "Gender", aggfunc = "sum")
foods.pivot_table(values = "Spend", index = "Item", aggfunc = "sum")
```

```python
foods.pivot_table(values = "Spend", index = ["Gender", "Item"], aggfunc = "sum")
```

```python
foods.pivot_table(values = "Spend", index = ["Gender", "Item"], columns = "City", aggfunc = "sum")
```

```python
foods.pivot_table(values = "Spend", index = ["Gender", "Item"], columns = ["Frequency", "City"] , aggfunc = "sum")
```

```python
foods.pivot_table(values = "Spend", index = ["Gender", "Item"], columns = "City" , aggfunc = "count")
foods.pivot_table(values = "Spend", index = ["Gender", "Item"], columns = "City" , aggfunc = "max")
foods.pivot_table(values = "Spend", index = ["Gender", "Item"], columns = "City" , aggfunc = "min")
```

```python
pd.pivot_table(data = foods, values = "Spend", index = ["Gender", "Item"], columns = "City" , aggfunc = "count")
```

# The pd.melt method():

- Reverse of what a pivot table does
- Called directly on pandas, not the df
- Takes an aggregated dataset and converts it to a tabular formula
- Hard to describe in words, but easier to understand visually

```python
sales = pd.read_csv("quarters.csv")
sales
```

```python
pd.melt(sales, id_vars = "Salesman")
# id_vars takes an arguement of a single column, indicates which column will be preserved
# here we've essentially condensed teh Q1, Q2, Q3, Q4 columns and reformatted the values to align with the columns
```

```python
pd.melt(sales, id_vars = "Salesman", var_name = "Quarter", value_name = "Revenue")
```
