# DF 3.2 (section 6.87):
## Retrieve Rows by Index Label with .loc[ ] accessor:

- .loc[ ] is used to access a row by index label

```python
bond = pd.read_csv("jamesbond.csv", index_col = "Film")
bond.sort_index(inplace = True)
# good habit to sort the index of a df before we start extracting any labels from it
# easier for pandas to find a value in a sorted collection than it is to find one in a jumbled one
bond.head(3)
```

```python
bond.loc["Goldfinger"]
# returns a series with the row's info displayed if only one row matches
# case sensitive
bond.loc["GoldenEye"]
```

```python
bond.loc["Casino Royale"]
# returns a df if multiple rows have the index being called
```

```python
bond.loc["Diamonds Are Forever":"From Russia with Love"]
# extracts the two rows called and all of the ones in between
# value after the colon is included
```

```python
bond.loc["Diamonds Are Forever":"From Russia with Love":2]
# the '2' is the step, skips every other movie in this example
# works the same as index slicing in python
```

```python
bond.loc["Goldeneye":]
bond.loc[:"Goldeneye"]
```

```python
# provide a list to extract more than one row label
bond.loc[["Dr. No", "For Your Eyes Only"]]
# order in the list determines the order they are outputted
fav_bond_movies = ["Die Another Day", "GoldenEye", "Casino Royale"]
bond.loc[fav_bond_movies]
# get a key error if any values dont exist in the df
```

```python
# check using the in operator
"GoldMember" in bond.index
```

# DF 3.4 (section 6.89):
## Second Arguements to .loc[ ] and .iloc[ ] Accessors:

- Customize which columns we are pulling out when we are accessing rows

```python
bond = pd.read_csv("jamesbond.csv", index_col = "Film")
bond.sort_index(inplace = True)
bond.head(3)
```

```python
bond.loc["Moonraker", "Actor"]
# put a second arguemnt inside the square brackets to specify the column, separated by a comnma
# returns the cell's value as a string
bond.loc["Moonraker", ["Actor", "Director"]]
# need a list to pull out mutiple columns or multiple indexes
# returns the result as a series
bond.loc["Moonraker":, ["Actor", "Director"]]
bond.loc[["GoldenEye", "Thunderball"], ["Actor", "Year"]]
```

```python
bond.loc["Moonraker", "Year":"Budget"]
# pulls out all columns from Year to Budget
bond.loc["From Russia with Love":"License to Kill", "Actor":"Budget"]
# right now, the indexes are in alphabetical order
```

```python
bond.iloc[16,2]
bond.iloc[3:6, 2:5]
# iloc works generally the same as loc, but you provide index positions and the last number is not included
```

# DF 3.6 (section 6.91):
## Set Multiple Values in a DataFrame:

```python
bond = pd.read_csv("jamesbond.csv", index_col = "Film")
bond.sort_index(inplace = True)
bond.head(3)
```

```python
actor_is_sean_connery = bond["Actor"] == "Sean Connery"
```

```python
# WRONG WAY TO DO THIS:
bond[actor_is_sean_connery]["Actor"] = "Sir Sean Connery"
# the issue: when we extract a subset using square brackets, we're creating a brand new df
# the values in the original df have not been affected
```

```python
# original bond has not been changed
bond
```

```python
# CORRECT WAY:
actor_is_sean_connery = bond["Actor"] == "Sean Connery"
bond.loc[actor_is_sean_connery, "Actor"] = "Bir Bean Bonnery"
# we aren't creating a copy here, we are extracting these rows
# can use .iloc[] in the same way if you want to use index positions
bond.head(9)
```

# DF 3.7 (section 6.92):
## Rename multiple index labels or columns in a dataframe:

```python
bond = pd.read_csv("jamesbond.csv", index_col = "Film")
bond.sort_index(inplace = True)
bond.head(3)
```

```python
bond.rename(mapper = {
    "GoldenEye": "Golden Eye", 
    "The Spy Who Loved Me": "The Spy Who Shagged Me"
    })
# can combine mapper and axis, or use index or columns independently
# give mapper a dictionary, the keys represent the existing labels in an axis, and the 
# values represent the new labels
# non-mutational, creates a brand new df
bond.rename(mapper = {
    "GoldenEye": "Golden Eye", 
    "The Spy Who Loved Me": "The Spy Who Shagged Me"
    }, axis = 0)
# ^ can provide an axis parameter to be more specific, axis = 0 means row, axis = 1 means column
```

```python
bond.rename(index = {
    "GoldenEye": "Golden Eye", 
    "The Spy Who Loved Me": "The Spy Who Shagged Me"
    }, inplace = True)
# easier, one less parameter
bond
```

```python
bond.head(1)
# renaming columns is the same logic as renaming rows (shown above), but use 'columnns' instead of 'index'
bond.rename(mapper = {"Year": "Release Year", "Box Office": "Revenue"}, axis = 1)
bond.rename(mapper = {"Year": "Release Year", "Box Office": "Revenue"}, axis = 'columns')
bond.rename(columns = {"Year": "Release Year", "Box Office": "Revenue"})
# ^ all three of these codes do the same thing ^
bond.rename(columns = {"Year": "Release Year", "Box Office": "Revenue"}, inplace = True)
bond.head(3)
```

```python
bond.columns = ["Year of release", "Actor", "Director", "Gross", "Cost", "Salary"]
# this is helpful when you want to rename all colunms at once, still useful if you want to rename a couple
# length of the list has to be identical to the length of the amount of columns
# have to provide a value for every column
# don't have to provide the former column names
bond.head(1)
```

# DF 3.9 (section 6.94):

# DF 3.11 (section 6.96):
## Filtering with the .where method:

- returns the original df, rows that are not represented are filled with null values
- good way to get back our original df but visually represent the data we want to look at

```python
bond = pd.read_csv("jamesbond.csv", index_col = "Film")
bond.sort_index(inplace = True)
bond.head(3)
```

```python
mask = bond["Actor"] == "Sean Connery"
bond.where(mask).head(5)
# accepts a boolean series as an arguement
```

```python
bond.where(bond["Box Office"] > 800).tail(4)
```

```python
# can feed multiple conditions into the where clause
mask1 = bond["Actor"] == "Sean Connery"
mask2 = bond["Box Office"] > 800
bond.where(mask1 & mask2).tail(4)
# have to use &, not 'and'
```

# DF 3.12 (section 6.97):
## the .query() Method:

- call directly on a df
- arguement must be a string
- column names can't have spaces

```python
bond = pd.read_csv("jamesbond.csv", index_col = "Film")
bond.sort_index(inplace = True)
bond.head(3)
```

```python
bond.columns = [column_name.replace(" ", "_") for column_name in bond.columns]
# need underscores instead of spaces to use the .query method
bond.head(1)
```

```python
bond.query('Actor == "Sean Connery"')
# faster than standard methods of extracting rows of larger df
```

```python
bond.query("Director != 'Terence Young'").head()
```

```python
bond.query('Box_Office > 600')
```

```python
bond.query("Actor == 'Roger Moore' and Director == 'John Glen'")
# multiple conditions syntax, use 'and', 'or' instead of &,| because its a python method, not a pandas method
```

```python
bond.query("Actor in ['Daniel Craig', 'Sean Connery', 'Roger Moore', 'Pierce Brosnan']")
bond.query("Actor not in ['Daniel Craig', 'Sean Connery', 'Roger Moore', 'Pierce Brosnan']")
```

# DF 3.13 (section 6.98):
## Review of .apply method on single columns:

- takes an existing python function and applies it to every value in a series


```python
bond = pd.read_csv("jamesbond.csv", index_col = "Film")
bond.sort_index(inplace = True)
bond.head(3)
```

```python
def convert_to_str_add_mil(number):
    return str(number) + ' Millions'

bond["Box Office"] = bond["Box Office"].apply(convert_to_str_add_mil)
```

```python
columns = ["Box Office", "Budget", "Bond Actor Salary"]

def convert_to_str_add_mil(number):
    return str(number) + ' Millions'

for col in columns:
    bond[col] = bond[col].apply(convert_to_str_add_mil)
    
bond.head(3)
```

# DF 3.14 (section 6.99):
## The .apply() method with row values:

- applies a single operation on a single column or series ( axis = 'rows')
- have to change axis to columns to apply the method to each row

```python
bond = pd.read_csv("jamesbond.csv", index_col = "Film")
bond.sort_index(inplace = True)
bond.head(3)
```

```python
def good_movie(row):
    
    actor = row[1]
    budget = row[4]
    
    if actor == "Pierce Brosnan":
        return "The Best"
    elif actor == "Roger Moore" and budget > 40:
        return "Enjoyable"
    else:
        return "IDK"
# This code defines a funciton that is expecting to receive a row  the df as its input, repeats this 
# code on every row that it's passed

bond.apply(good_movie, axis = "columns")
# axis is set to columns even though you want it to read rows because you want it to 'move across' the rows
```

# DF 3.15 (section 6.100):
## The .copy() method:

- creates an exact copy of an existing pandas object (such as a df or series)
- stores the copy completely separately in memory
- changes to the copy won't affect the original df

```python
bond = pd.read_csv("jamesbond.csv", index_col = "Film")
bond.sort_index(inplace = True)
bond.head(3)
```

```python
directors = bond["Director"].copy()
directors.head(3)
```

```python
directors["A View to a Kill"] = "Mister John Glen"
directors.head(3)
```

```python
bond.head(3)
# original df is unchanged
```
